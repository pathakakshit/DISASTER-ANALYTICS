```python
#lIBRARY USED IN THIS PROJECT
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#DATASET
df = pd.read_excel('GROUP_ET_PFM_2.xlsx')
df
```

```python
#META DATA
df.columns
```

###### Before getting into data analytics let us try to find is there any null values or missing data in this data set and find the summary statistics of the data to get an idea about the dataset.

```python
df.update(df[['Total Deaths','No. Injured','No. Affected','No. Homeless','Total Affected',
              'Reconstruction Costs','Insured Damage','Total Damage','CPI']].fillna(0))
```

Here, to treat null values, a concept of Constant Value Imputation is used. As 'Total Deaths','No. Injured','No. Affected','No. Homeless','Total Affected',
              'Reconstruction Costs','Insured Damage','Total Damage','CPI' are the absolute values and null here means zero. Therefore null values are been filled with 0.

```python
df.drop(columns=['Origin','Associated Types','AID Contribution','Magnitude'], inplace=True)
```

Columns which have no significance in the future analyses are dropped from the data set

```python
df.isnull().sum()
```

# FINDINGS

###### 1. Number of total deaths with respect to the disaster type

```python
df.groupby("Disaster Type")["Total Deaths"].sum().plot(figsize=(12,5),
kind="bar", xlabel="Disaster_Type", ylabel="Number of deaths", title="Deaths by Disaster Types", logy=True)
```

Interpretation: Here we can see that the most number of deaths has been due to Drought. The govenment should implement policies to deal with this problem. 

###### 2. Number of total deaths with respect to the Subregion

```python
df.groupby('Subregion')['Total Deaths'].sum().sort_values()
```

Interpretation: Here we can see that the most number of deaths has been seen in Eastern Asia, while the lowest number of deaths has been shown in Micronesia.

###### 3. Relaionship between CPI and Total number of deaths, grouped by Subregion

```python
import pandas as pd
import matplotlib.pyplot as plt
result = df.groupby('Subregion').agg({'Total Deaths': 'sum', 'CPI': 'mean'}).reset_index()

print(result)

plt.figure(figsize=(10, 6))
plt.bar(result['Subregion'], result['Total Deaths'], label='Total Deaths', color='red')
plt.plot(result['Subregion'], result['CPI'], label='Average CPI', color='blue', marker='o')
plt.xlabel('Subregion')
plt.ylabel('Count / Index')
plt.title('Total Deaths and Average CPI by Subregion')
plt.legend()
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

Interpretation: There is a negative correlation between the total number of deaths and the average CPI Index. High average CPI have low death numbers, as they are already prepared for any natural disaster.

###### 4. Deaths by Disaster Subtypes in Asia

```python
df[df["Region"]=="Asia"].groupby(["Disaster Subtype"])["Total Deaths"].sum().plot(figsize=(15,5), ylabel="Casualties",
logy=False, title="Deaths by Disaster Subtypes in Asia")
plt.show()
```

###### 5. Top 5 countries with the highest total affected population

```python
top_affected_countries = df.groupby('Country')['Total Affected'].sum().nlargest(5)
print(f"4. Top 5 countries with the highest total affected population:\n{top_affected_countries}")
```

###### 6. Top 10 Countries with Highest Total Damage

```python
top_countries_damage = df.groupby('Country')['Total Damage'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_countries_damage.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Highest Total Damage')
plt.xlabel('Country')
plt.ylabel('Total Damage')
plt.show()
```

###### 7. Distribution of Disaster Types

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='Disaster Type', data=df, order=df['Disaster Type'].value_counts().index, palette='muted')
plt.title('Distribution of Disaster Types')
plt.xlabel('Number of Occurrences')
plt.ylabel('Disaster Type')
plt.show()
```

###### 8. Number of Disaster Deaths by Region

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.barplot(x='Region', y='Total Deaths', data=df, palette='viridis', ci=None)
plt.title('Number of Disaster Deaths by Region')
plt.xlabel('Region')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.show()
```

###### 9. Trend in Disaster Occurrences over Time

```python
disasters_per_year = df.groupby('Start Year').size()
 
plt.figure(figsize=(10, 6))
disasters_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Disasters')
plt.title('Trend in Disaster Occurrences over Time')
plt.xticks(rotation=45)
plt.show()
```
